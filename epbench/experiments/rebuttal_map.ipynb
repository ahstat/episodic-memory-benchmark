```python
import folium
from folium.raster_layers import TileLayer
from folium.features import DivIcon

main_entity = 'Jackson Ramos' # good name, because all the elements are different (otherwise difficult to plot), and also all in New York center
# Define locations
locations = [{'name': k, 'coords': [v[1], v[0]], 'color': '#FF5733'} for k, v in location_to_position.items()]

# Create a map centered on New York City
nyc_map = folium.Map(
    location=[40.7128, -74.0060],  # Center of NYC
    zoom_start=12,
    tiles=None,
)

# Add minimal light tiles
TileLayer(
    tiles='https://{s}.basemaps.cartocdn.com/light_nolabels/{z}/{x}/{y}{r}.png',
    attr='&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors &copy; <a href="https://carto.com/attributions">CARTO</a>',
    name='minimal',
    control=False,
    opacity=0.7
).add_to(nyc_map)

from folium import plugins

# Add path markers for each location
for i, location in enumerate(locations, 1):

    if location['name'] in [x[1] for x in all_events_with_coordinates]:
        folium.CircleMarker(
            location=location['coords'],
            radius=3,
            color='#666666',
            fill=True,
            fill_opacity=1,
            weight=0
        ).add_to(nyc_map)

already_done = []
for k,v in per_entity_all_events_with_coordinates.items():

    if k == main_entity:
        color='#FF6666'
        opacity=0.7
    else:
        color='#666666'
        opacity=0.1

    line_coordinates = [x[5] for x in v]
    my_time = [x[0] for x in v]
    my_location = [x[1] for x in v]
    my_entity = [x[2] for x in v]
    my_content = [x[3] for x in v]
    my_detail = [x[4] for x in v]
    my_position = [x[5] for x in v]
    my_chapter = [x[6] for x in v]



    for i in range(len(line_coordinates) - 1):
        start = line_coordinates[i]
        end = line_coordinates[i + 1]

        if (('/'.join([str(x) for x in start + end])) not in already_done) or (k == main_entity):
            already_done.append('/'.join([str(x) for x in start + end]))
            #print(already_done)
        
            # Create the line with arrow markers
            folium.PolyLine(
                locations=[start, end],
                weight=2,
                color=color,
                opacity=opacity,
                colormap=["y", "orange", "r"],
                #arrow_style='->',  # Add arrow to the line
                #arrow_size=15,
            ).add_to(nyc_map)

            # Add smaller bullet point
            if k == main_entity:
                folium.CircleMarker(
                    location=start,
                    radius=3,
                    color='#FF6666',
                    fill=True,
                    fill_opacity=1,
                    weight=0
                ).add_to(nyc_map)

                # Add annotation bubble
                folium.Marker(
                    location=start,
                    icon=DivIcon(
                        icon_size=(338,36),
                        icon_anchor=(0,0), # font-weight: bold;
                        html=f"""
                        <div style="
                            background-color: #fff; 
                            color: #333;
                            font-size: 14px;
                            padding: 4px 6px;
                            border: 1px solid #ccc;
                            white-space: nowrap;
                        "><center>
                            {my_chapter[i]} ({my_time[i]})<br>
                            at {my_location[i]} 
                            doing {my_content[i]}
                        </center></div>
                        """
                    )
                ).add_to(nyc_map)
            
            if i == (len(line_coordinates) - 2): # last segment, so also consider the end
                if k == main_entity:
                    folium.CircleMarker(
                        location=end,
                        radius=3,
                        color='#FF6666',
                        fill=True,
                        fill_opacity=1,
                        weight=0
                    ).add_to(nyc_map)
                    # Add annotation bubble
                    folium.Marker(
                        location=end,
                        icon=DivIcon(
                            icon_size=(338,36),
                            icon_anchor=(0,0), #font-weight: bold;
                            html=f"""
                            <div style="
                                background-color: #fff; 
                                color: #333;
                                font-size: 14px;
                                padding: 4px 6px;
                                border: 1px solid #ccc;
                                white-space: nowrap;
                            "><center>
                                {my_chapter[i+1]} ({my_time[i+1]})<br>
                                at {my_location[i+1]} 
                                doing {my_content[i+1]}
                            </center></div>
                            """
                        )
                    ).add_to(nyc_map)

# Save the map to an HTML file
nyc_map.save(Path(git_repo_filepath) / 'epbench' / 'plots' / f'nyc_minimal_map.html')
```
