```python
from epbench.src.results.average_groups import extract_groups
import pandas as pd

# post-processing of tables for visual rendering
def reshaping_dictionary_of_dataframes(df_results_book, nb_events):
    if nb_events == 20:
        book_size = 'short'
    else:
        book_size = 'long'

    # First, rename the 'c' column in each DataFrame to c1, c2, c3
    renamed_dfs = {}
    for key, df in df_results_book.items():
        df_copy = df.copy()
        df_copy = df_copy.rename(columns={('prompting', 'gpt-4o-2024-08-06', 'n/a'): f"{book_size} {key}"})
        renamed_dfs[key] = df_copy

    # Merge all DataFrames on 'bins_items_correct_answer
    result = renamed_dfs[list(renamed_dfs.keys())[0]][['bins_items_correct_answer']].copy()

    # Add the results columns
    for key, df in renamed_dfs.items():
        result = result.merge(df[['bins_items_correct_answer', f"{book_size} {key}"]], on='bins_items_correct_answer', how='left')

    # Reorder columns if needed
    result = result[['bins_items_correct_answer'] + [f"{book_size} {key}" for key in renamed_dfs.keys()]]

    return result

res_final = {}
for nb_events in [20, 200]:
    # nb_events = 200 # select the book of interest (either 20 or 200)
    relative_to = ['get', 'bins_items_correct_answer'] # select the grouped elements as a list among:
    # 'get': type of question, among 'all' (simple recall questions), 'latest' (latest state questions), or 'chronological' (chronological questions)
    # 'bins_items_correct_answer': number of events for this question, binned into {0}, {1}, {2}, {3,4,5}, {6+} chapters
    # 'cue': type of cue for this question, e.g. (*,*,*,c)
    # 'retrieval_type': type of trace for this question, e.g. 'Spaces'
    book = 'scifi'
    df_results_book = {}
    for book in ['default', 'news', 'scifi']:
        df_results = extract_groups(df[df['book'] == book], nb_events, relative_to) # group the results according to `relative_to`
        # Further filtering, e.g. for selecting only the simple recall questions:
        df_results = df_results[df_results['get'] == 'all'].drop('get', axis = 1)
        df_results_book[book] = df_results

    res = reshaping_dictionary_of_dataframes(df_results_book, nb_events).transpose()
    res_final[nb_events] = res.set_axis(res.iloc[0], axis=1).iloc[1:]

pd.concat([res_final[20], res_final[200]])
```
