### 3. Comparing pairs of experiments

```python
import numpy as np
import pandas as pd
import dabest
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------
# 1. Load and Rename the Data
# ---------------------

# Load the data
df = pd.read_csv(Path(git_repo_filepath) / 'data' / 'result_lenient_all_book_200.csv')

# Rename columns for clarity
df.columns = [
    'ID',
    'ppt/min',      # Prompting method, small group
    'ppt/4o',       # Prompting method, large group
    'ppt/hai',      # Prompting method, small group
    'ppt/s35',      # Prompting method, large group
    'rg/min/p',     # RAG method, small group, paragraph
    'rg/4o/p',      # RAG method, large group, paragraph
    'rg/hai/p',     # RAG method, small group, paragraph
    'rg/s35/p',     # RAG method, large group, paragraph
    'rg/min/c',     # RAG method, small group, chapter
    'rg/4o/c',      # RAG method, large group, chapter
    'rg/hai/c',     # RAG method, small group, chapter
    'rg/s35/c'      # RAG method, large group, chapter
]

# Verify the renamed columns
print("Renamed Columns:", df.columns.tolist())

# ---------------------
# 2. Define Groupings for Multi-Baseline Repeated Measures Analysis
# ---------------------

# Define groupings for multi-baseline repeated measures (mutually exclusive)
multi_groups = {
    'Prompting Small': ['ppt/min', 'ppt/hai'],
    'Prompting Large': ['ppt/4o', 'ppt/s35'],
    'RAG Small Paragraph': ['rg/min/p', 'rg/hai/p'],
    'RAG Large Paragraph': ['rg/4o/p', 'rg/s35/p'],
    'RAG Small Chapter': ['rg/min/c', 'rg/hai/c'],
    'RAG Large Chapter': ['rg/4o/c', 'rg/s35/c']
}

# ---------------------
# 3. Compute and Sort Groups by Average Scores
# ---------------------

# Compute average scores for each group
group_averages = {}
for group_name, columns in multi_groups.items():
    # Compute the mean across the specified columns for each row, then take the overall mean
    group_averages[group_name] = df[columns].mean(axis=1).mean()

# Sort groups by average score in descending order (best to worst)
sorted_multi_groups = sorted(multi_groups.items(), key=lambda x: group_averages[x[0]], reverse=True)

# Display sorted groups
print("\nMulti-Baseline Groups sorted by average scores (best to worst):")
for group_name, _ in sorted_multi_groups:
    print(f"{group_name}: {group_averages[group_name]:.4f}")

# ---------------------
# 4. Prepare `idx` for Multi-Baseline Repeated Measures Analysis
# ---------------------

# Prepare idx tuples for multi-baseline repeated measures (ONLY column names)
multi_idx = tuple([
    tuple(columns) for group_name, columns in sorted_multi_groups
])

# Display the idx structure
print("\nMulti-Baseline DABEST idx structure:")
for item in multi_idx:
    print(item)

# ---------------------
# 5. Load Data into DABEST for Multi-Baseline Repeated Measures Analysis
# ---------------------

# Load data into DABEST for multi-baseline repeated measures analysis
multi_baseline_repeated_measures = dabest.load(
    data=df,
    idx=multi_idx,
    paired='baseline',  # Indicates that measurements are paired based on the baseline
    id_col='ID'         # Column that uniquely identifies each subject or unit
)

# ---------------------
# 6. Define and Prepare RAG Chapter vs. Paragraph Comparisons Separately
# ---------------------

# Define RAG Chapter vs. Paragraph comparisons
rag_comparisons = {
    'RAG Small: Paragraph vs. Chapter': ['rg/min/p', 'rg/min/c', 'rg/hai/p', 'rg/hai/c'],
    'RAG Large: Paragraph vs. Chapter': ['rg/4o/p', 'rg/4o/c', 'rg/s35/p', 'rg/s35/c']
}

# Compute average scores for RAG comparisons
rag_group_averages = {}
for group_name, columns in rag_comparisons.items():
    rag_group_averages[group_name] = df[columns].mean(axis=1).mean()

# Sort RAG comparison groups by average score
sorted_rag_comparisons = sorted(rag_comparisons.items(), key=lambda x: rag_group_averages[x[0]], reverse=True)

# Prepare idx tuples for RAG Chapter vs. Paragraph comparisons (ONLY column names)
rag_idx = tuple([
    tuple(columns) for group_name, columns in sorted_rag_comparisons
])

# Display the rag_idx structure
print("\nRAG Chapter vs. Paragraph DABEST idx structure:")
for item in rag_idx:
    print(item)

# Load data into DABEST for RAG Chapter vs. Paragraph comparisons
rag_repeated_measures = dabest.load(
    data=df,
    idx=rag_idx,
    paired='baseline',  # Indicates that measurements are paired based on the baseline
    id_col='ID'         # Column that uniquely identifies each subject or unit
)

# ---------------------
# 7. Identify Best Groups for Targeted Comparisons
# ---------------------

# Identify best RAG group
rag_groups = {k: v for k, v in group_averages.items() if 'RAG' in k}
best_rag_group = max(rag_groups, key=rag_groups.get)

# Identify best Prompting group
prompting_groups = {k: v for k, v in group_averages.items() if 'Prompting' in k}
best_prompting_group = max(prompting_groups, key=prompting_groups.get)

print(f"\nBest RAG Group: {best_rag_group} with average score {rag_groups[best_rag_group]:.4f}")
print(f"Best Prompting Group: {best_prompting_group} with average score {prompting_groups[best_prompting_group]:.4f}")

# Identify best Small group
small_groups = {k: v for k, v in group_averages.items() if 'Small' in k}
best_small_group = max(small_groups, key=small_groups.get)

# Identify best Large group
large_groups = {k: v for k, v in group_averages.items() if 'Large' in k}
best_large_group = max(large_groups, key=large_groups.get)

print(f"Best Small Group: {best_small_group} with average score {small_groups[best_small_group]:.4f}")
print(f"Best Large Group: {best_large_group} with average score {large_groups[best_large_group]:.4f}")

# ---------------------
# 8. Create Comparison DataFrames with Concise Labels
# ---------------------

# Helper function to create comparison DataFrames
def create_comparison_df(group1_columns, group2_columns, group1_label, group2_label):
    """
    Create a comparison DataFrame with concise labels.

    Parameters:
    - group1_columns: list of column names for group 1
    - group2_columns: list of column names for group 2
    - group1_label: string label for group 1
    - group2_label: string label for group 2

    Returns:
    - pandas DataFrame with averaged scores and ID
    """
    df_group1 = df[group1_columns].mean(axis=1)
    df_group2 = df[group2_columns].mean(axis=1)
    comparison_df = pd.DataFrame({
        group1_label: df_group1,
        group2_label: df_group2,
        'ID': df['ID']
    })
    return comparison_df

# a. Best RAG vs. Best Prompting
comparison_df_rag_prompting = create_comparison_df(
    multi_groups[best_rag_group],
    multi_groups[best_prompting_group],
    'best/r/p',   # best RAG Paragraph
    'best/ppt'    # best Prompting
)

# b. Best Small vs. Best Large
comparison_df_small_large = create_comparison_df(
    multi_groups[best_small_group],
    multi_groups[best_large_group],
    'best/small',
    'best/large'
)

# c. Best RAG vs. Best Small
comparison_df_rag_small = create_comparison_df(
    multi_groups[best_rag_group],
    multi_groups[best_small_group],
    'best/r/p',
    'best/small'
)

# d. Best RAG vs. Best Large
comparison_df_rag_large = create_comparison_df(
    multi_groups[best_rag_group],
    multi_groups[best_large_group],
    'best/r/p',
    'best/large'
)

# e. Best Prompting vs. Best Small
comparison_df_ppt_small = create_comparison_df(
    multi_groups[best_prompting_group],
    multi_groups[best_small_group],
    'best/ppt',
    'best/small'
)

# f. Best Prompting vs. Best Large
comparison_df_ppt_large = create_comparison_df(
    multi_groups[best_prompting_group],
    multi_groups[best_large_group],
    'best/ppt',
    'best/large'
)

# ---------------------
# 9. Load Data into DABEST for Comparisons
# ---------------------

# Function to load DABEST object
def load_dabest(comparison_df, group1_label, group2_label):
    """
    Load a DABEST object for a given comparison.

    Parameters:
    - comparison_df: pandas DataFrame containing the comparison data
    - group1_label: string label for group 1
    - group2_label: string label for group 2

    Returns:
    - DABEST object
    """
    dabest_obj = dabest.load(
        data=comparison_df,
        idx=(group1_label, group2_label),
        paired='baseline',
        id_col='ID'
    )
    return dabest_obj

# a. Load Best RAG vs. Best Prompting
best_rag_vs_prompting = load_dabest(
    comparison_df_rag_prompting,
    'best/r/p',
    'best/ppt'
)

# b. Load Best Small vs. Best Large
best_small_vs_large = load_dabest(
    comparison_df_small_large,
    'best/small',
    'best/large'
)

# c. Load Best RAG vs. Best Small
best_rag_vs_small = load_dabest(
    comparison_df_rag_small,
    'best/r/p',
    'best/small'
)

# d. Load Best RAG vs. Best Large
best_rag_vs_large = load_dabest(
    comparison_df_rag_large,
    'best/r/p',
    'best/large'
)

# e. Load Best Prompting vs. Best Small
best_ppt_vs_small = load_dabest(
    comparison_df_ppt_small,
    'best/ppt',
    'best/small'
)

# f. Load Best Prompting vs. Best Large
best_ppt_vs_large = load_dabest(
    comparison_df_ppt_large,
    'best/ppt',
    'best/large'
)

# ---------------------
# 10. Define a Reusable Function for Comparisons
# ---------------------

def compare_and_plot(dabest_obj, group1_label, group2_label, output_filename):
    """
    Compare two groups and plot the mean differences using DABEST with enhanced plotting parameters.

    Parameters:
    - dabest_obj: DABEST object for the comparison
    - group1_label: string label for the first group (shortened)
    - group2_label: string label for the second group (shortened)
    - output_filename: string for saving the plot

    Returns:
    - None
    """
    # Create Gardner-Altman plot with enhanced parameters
    fig = dabest_obj.mean_diff.plot(
        fig_size=(5, 4),  # Adjusted figure size for better visibility
        dpi=150,
        reflines_kwargs={
            'linestyle': 'dashed',
            'linewidth': 0.8,
            'color': 'black'
        },
        swarm_label=None,       # Remove or customize as needed
        show_pairs=True,        # Show individual paired observations
        swarm_desat=1,
        group_summaries='mean_sd',
        swarmplot_kwargs={'size':5},
        group_summary_kwargs={'lw':3, 'alpha':0.8},
        float_contrast=True,
        contrast_label='paired mean difference',
        es_marker_size=5,
        halfviolin_desat=1,
        halfviolin_alpha=0.8,
        violinplot_kwargs={'widths':0.5}
    )
    
    # Rotate x-tick labels by 45 degrees for better readability
    for ax in fig.axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    
    # Adjust layout to prevent clipping of tick-labels
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(output_filename, dpi=150)
    
    # Display the plot
    plt.show()

# ---------------------
# 11. Generate Targeted Plots
# ---------------------

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# a. Plot Best RAG vs. Best Prompting
compare_and_plot(
    dabest_obj=best_rag_vs_prompting,
    group1_label='best/r/p',
    group2_label='best/ppt',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_rag_vs_prompting.png'
)

# b. Plot Best Small vs. Best Large
compare_and_plot(
    dabest_obj=best_small_vs_large,
    group1_label='best/small',
    group2_label='best/large',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_small_vs_large.png'
)

# c. Plot Best RAG vs. Best Small
compare_and_plot(
    dabest_obj=best_rag_vs_small,
    group1_label='best/r/p',
    group2_label='best/small',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_rag_vs_small.png'
)

# d. Plot Best RAG vs. Best Large
compare_and_plot(
    dabest_obj=best_rag_vs_large,
    group1_label='best/r/p',
    group2_label='best/large',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_rag_vs_large.png'
)

# e. Plot Best Prompting vs. Best Small
compare_and_plot(
    dabest_obj=best_ppt_vs_small,
    group1_label='best/ppt',
    group2_label='best/small',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_ppt_vs_small.png'
)

# f. Plot Best Prompting vs. Best Large
compare_and_plot(
    dabest_obj=best_ppt_vs_large,
    group1_label='best/ppt',
    group2_label='best/large',
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'best_ppt_vs_large.png'
)

# ---------------------
# 12. Plot Multi-Baseline Repeated Measures Analysis
# ---------------------

# Function to plot multi-baseline comparisons
def plot_multi_baseline(dabest_obj, output_filename):
    """
    Plot multi-baseline repeated measures analysis.

    Parameters:
    - dabest_obj: DABEST object for multi-baseline analysis
    - output_filename: string for saving the plot

    Returns:
    - None
    """
    # Create multi-baseline plot
    fig = dabest_obj.mean_diff.plot(
        fig_size=(14, 8),  # Increased size for multiple comparisons
        dpi=150,
        reflines_kwargs={
            'linestyle': 'dashed',
            'linewidth': 0.8,
            'color': 'black'
        },
        swarm_label=None,        # Remove or customize as needed
        show_pairs=False,        # Pairs are already considered in multi-baseline
        swarm_desat=1,
        group_summaries='mean_sd',
        swarmplot_kwargs={'size':5},
        group_summary_kwargs={'lw':3, 'alpha':0.8},
        float_contrast=True,
        contrast_label='paired mean difference',
        es_marker_size=5,
        halfviolin_desat=1,
        halfviolin_alpha=0.8,
        violinplot_kwargs={'widths':0.5}
    )
    
    # Rotate x-tick labels by 45 degrees for better readability
    for ax in fig.axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    
    # Adjust layout to prevent clipping of tick-labels
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(output_filename, dpi=150)
    
    # Display the plot
    plt.show()

# Plot Multi-Baseline Repeated Measures
plot_multi_baseline(
    dabest_obj=multi_baseline_repeated_measures,
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'multi_baseline_repeated_measures_analysis.png'
)

# ---------------------
# 13. Plot RAG Chapter vs. Paragraph Comparisons Separately
# ---------------------

# Function to plot RAG Chapter vs. Paragraph comparisons
def plot_rag_comparisons(dabest_obj, output_filename):
    """
    Plot RAG Chapter vs. Paragraph comparisons.

    Parameters:
    - dabest_obj: DABEST object for RAG comparisons
    - output_filename: string for saving the plot

    Returns:
    - None
    """
    # Create plot
    fig = dabest_obj.mean_diff.plot(
        fig_size=(10, 6),  # Adjusted size for better visibility
        dpi=150,
        reflines_kwargs={
            'linestyle': 'dashed',
            'linewidth': 0.8,
            'color': 'black'
        },
        swarm_label=None,        # Remove or customize as needed
        show_pairs=False,        # Show or hide based on preference
        swarm_desat=1,
        group_summaries='mean_sd',
        swarmplot_kwargs={'size':5},
        group_summary_kwargs={'lw':3, 'alpha':0.8},
        float_contrast=True,
        contrast_label='paired mean difference',
        es_marker_size=5,
        halfviolin_desat=1,
        halfviolin_alpha=0.8,
        violinplot_kwargs={'widths':0.5}
    )
    
    # Rotate x-tick labels by 45 degrees for better readability
    for ax in fig.axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    
    # Adjust layout to prevent clipping of tick-labels
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(output_filename, dpi=150)
    
    # Display the plot
    plt.show()

# Plot RAG Chapter vs. Paragraph Comparisons
plot_rag_comparisons(
    dabest_obj=rag_repeated_measures,
    output_filename=Path(git_repo_filepath) / 'plots' / 'gardner_altman' / 'rag_chapter_vs_paragraph_analysis.png'
)

# ---------------------
# 14. Additional Semantic Comparisons (Optional)
# ---------------------

# Example: Best RAG Paragraph vs. Best RAG Chapter
def plot_additional_comparison(dabest_obj, group1_label, group2_label, output_filename):
    """
    Plot an additional semantic comparison.

    Parameters:
    - dabest_obj: DABEST object for the comparison
    - group1_label: string label for group 1
    - group2_label: string label for group 2
    - output_filename: string for saving the plot

    Returns:
    - None
    """
    # Create plot
    fig = dabest_obj.mean_diff.plot(
        fig_size=(5, 4),
        dpi=150,
        reflines_kwargs={
            'linestyle': 'dashed',
            'linewidth': 0.8,
            'color': 'black'
        },
        swarm_label=None,
        show_pairs=True,
        swarm_desat=1,
        group_summaries='mean_sd',
        swarmplot_kwargs={'size':5},
        group_summary_kwargs={'lw':3, 'alpha':0.8},
        float_contrast=True,
        contrast_label='paired mean difference',
        es_marker_size=5,
        halfviolin_desat=1,
        halfviolin_alpha=0.8,
        violinplot_kwargs={'widths':0.5}
    )
    
    # Rotate x-tick labels by 45 degrees for better readability
    for ax in fig.axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='center')
    
    # Adjust layout to prevent clipping of tick-labels
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(output_filename, dpi=150)
    
    # Display the plot
    plt.show()
```
